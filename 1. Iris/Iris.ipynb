# Assignment 1: Iris Flower Dataset Exploratory Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_df = pd.read_csv("data/iris.data.csv")
```

```python
iris_df.head()
```

```python
iris_df.columns = ['sepal_length', 'sepal_width' , 'petal_length', 'petal_width', 'species']
iris_df.head()
```

## 1. How many features are there and what are their types (e.g. numeric, nominal)?

```python
print(iris_df.dtypes)
```

##  2. Compute and display summary statistics for each feature available in the dataset.

```python
summary = iris_df.describe()
summary = summary.transpose()
summary.head()
```

## 3. Data Visualization - Create a histogram for each feature available in the dataset.

```python
print(iris_df.hist())
```

```python
sns.pairplot(iris_df, hue="species", palette="Set2", markers=["o", "s", "D"])
plt.show()
```

## 4.1. Create a boxplot for each feature in the dataset.

```python
sns.set(style="whitegrid", palette="Set2", rc={'figure.figsize':(11.7,8.27)})
title="Compare the Distributions of Sepal Length"
sns.boxplot(x="species", y="sepal_length", data=iris_df)
plt.title(title, fontsize=26)
plt.show()
```

```python
sns.set(style="whitegrid", palette="Set2", rc={'figure.figsize':(11.7,8.27)})
title="Compare the Distributions of Sepal Width"
sns.boxplot(x="species", y="sepal_width", data=iris_df)
plt.title(title, fontsize=26)
plt.show()
```

```python
sns.set(style="whitegrid", palette="Set2", rc={'figure.figsize':(11.7,8.27)})
title="Compare the Distributions of Petal Length"
sns.boxplot(x="species", y="petal_length", data=iris_df)
plt.title(title, fontsize=26)
plt.show()
```

```python
sns.set(style="whitegrid", palette="Set2", rc={'figure.figsize':(11.7,8.27)})
title="Compare the Distributions of Petal Width"
sns.boxplot(x="species", y="petal_width", data=iris_df)
plt.title(title, fontsize=26)
plt.show()
```

## 4.2. All of the boxplots should be combined into a single plot. Compare distributions and identify outliers.

```python
iris_df.plot(kind = 'box')
plt.ylabel('[cm]')
plt.show()
```
